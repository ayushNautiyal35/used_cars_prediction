```python
d={}
def dictionary_convert(i):
    d[i]={'other':0}
    k=1
    for j in data[i].unique():
        d[i][j]=k
        k+=1
    
    
```

```python
import numpy as np

q1=data['milage'].quantile(0.25)
q3=data['milage'].quantile(0.75)
iqr=q3-q1
maximum=q3+1.5*iqr
minimum=q1-1.5*iqr
data['milage']=np.where(data['milage']>=maximum,data['milage'].median(),data['milage'])
data['milage']=np.where(data['milage']<=minimum,data['milage'].median(),data['milage'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, target, random_state=42)
```

```python
import math
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
def check(model):
    y_pred=model.predict(x_test)
    print("Mean _sqare_error",mean_squared_error(y_test,y_pred))
    print("Mean absolute error",mean_absolute_error(y_test,y_pred))
    print("r2 score",r2_score(y_test,y_pred))
    print("rmse",math.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.linear_model import LinearRegression
check(LinearRegression().fit(x_train,y_train))
```

```python
from sklearn.neighbors import KNeighborsRegressor
check(KNeighborsRegressor().fit(x_train,y_train))
```

```python
from sklearn.tree import DecisionTreeRegressor
check(DecisionTreeRegressor().fit(x_train,y_train))
```

```python
from sklearn.ensemble import RandomForestRegressor
check(RandomForestRegressor().fit(x_train,y_train))
```

```python
from sklearn.ensemble import AdaBoostRegressor
check(AdaBoostRegressor().fit(x_train,y_train))
```

```python
from sklearn.ensemble import BaggingRegressor
check(BaggingRegressor().fit(x_train,y_train))
```

```python
from sklearn.ensemble import ExtraTreesRegressor
check(ExtraTreesRegressor().fit(x_train,y_train))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
check(GradientBoostingRegressor().fit(x_train,y_train))
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
check(HistGradientBoostingRegressor().fit(x_train,y_train))
```

```python
ml_model=HistGradientBoostingRegressor().fit(x_train,y_train)
ml_model
```
